```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Convolution2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Activation
from keras import backend as K
```

```python
class LeNet:
    @staticmethod
    def build(width, height, depth, classes):
        model = Sequential()
        inputShape = (height, width, depth)

        if K.image_data_format() == 'channels_first':
            inputShape = (depth, height, width)

        model.add(Convolution2D(20, (5, 5), padding='SAME', input_shape=inputShape))
        model.add(Activation('relu'))
        model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))

        model.add(Convolution2D(50, (5, 5), padding='SAME'))
        model.add(Activation('relu'))
        model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))

        model.add(Flatten())

        # Adjust the number of units in the Dense layer to match the Flatten layer's output size
        model.add(Dense(3200))  # You may need to adjust this based on your model's output size
        model.add(Activation('relu'))

        # Change the number of units in the last Dense layer for binary classification
        model.add(Dense(2))
        model.add(Activation('softmax'))

        return model
```
